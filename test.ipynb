```python
import numpy as np
import tushare as ts
from tqdm import tqdm
import pandas as pd

ts.set_token("e95d6e0dc53b83f930b93148ce9138f11bd6ae0e2550b02f6a3798aa")
pro = ts.pro_api()
```

```python
df = pro.cn_pmi(start_m='201601', end_m='202606')
```

```python
names = pro.stock_basic(exchange = '', list_status = 'L', fields = 'ts_code, symbol, name, area, industry, list_date')
```

```python
codes = names["ts_code"].values
master_df = []
```

```python
start = len(master_df)
for code in tqdm(codes[start : ]):
    df = pro.daily(ts_code = code, start_date = '20160601', end_date = '20260601')
    master_df.append(df)
```

```python
for i, df in enumerate(master_df[start: ]): 
    df.to_csv(f"data/daily_prices/{codes[i + start]}.csv", index = False)
```
